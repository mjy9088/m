# 신경망

신경망은 퍼셉트론과 유사하나, 퍼셉트론은 가중치를 사람이 직접 설정해야 하지만

신경망은 적절한 가중치 값을 데이터로부터 자동으로 학습할 수 있음

## 퍼셉트론과의 차이

뉴런이 연결되는 방식은 퍼셉트론과 같음

퍼셉트론은 0 또는 1을 출력할 수 있음 (_각 입력과 가중치의 곱의 합_이 임계값을 넘으면 1을 출력)

뉴런은 _각 입력과 가중치의 곱의 합_과 편향을 더한 값으로 출력할 값을 정하는 **활성화 함수**가 있음

입력 신호가 들어가는 층을 **입력층**, 출력이 나오는 마지막 층을 **출력층**, 그 중간의 나머지 층을 **은닉층**이라고 함

## 활성화 함수

입력 신호의 총합을 출력 신호로 변환하는 함수

```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)
plt.plot(x, y);
plt.ylim(-0.1, 1.1);
plt.show()
```

위는 계단 함수의 그래프로, 퍼셉트론이 활성화함수로 계단함수를 사용함.

신경망은 활성화 함수를 주로 계단 함수가 아닌 다른 함수로 사용함

아래는 활성화 함수 예시 sigmoid의 그래프

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y);
plt.ylim(-0.1, 1.1);
plt.show()
```

신경망의 활성화 함수로는 비선형 함수를 사용해야 함 (선형 함수는 f(x) = ax + b 꼴)

신경망의 활성화 함수로 선형 함수를 사용하면 은닉층을 갖는 의미가 없어짐 (층 하나로 모두 표현 가능)

아래는 또 다른 활성화 함수 예시 ReLU의 그래프

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y);
plt.show()
```
